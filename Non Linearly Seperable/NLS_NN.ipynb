```python
import pandas as pd 
import numpy as np 
import KNN
```

```python
train=pd.read_csv('train.csv')
test=pd.read_csv('test.csv')
```

```python
X=train[['x1','x2']].values
y=train['Class'].values
```

```python
knn=KNN.KNN(k=1)
knn.fit(X,y)
knn.plot_decision_boundary(X, y)
```

```python
x=test[['x1','x2']].values
y_true=test['Class'].values
```

```python
final_pred=pd.DataFrame(x,columns=['x1','x2'])
final_pred['y_cap']=knn.predict(x)
final_pred
```

```python
f=final_pred[['x1','x2']].values
y_hat=final_pred['y_cap'].values
```

```python
import accuracy
acc=accuracy.accuracy(y_true,y_hat)
```

```python
acc.confusion_matrix()
```

```python
acc.accu_percent()
```

```python
acc.precision()
```

```python
acc.recall()
```

```python
acc.f_score()
```
