```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_dataset(FilePath):
    data=np.loadtxt(FilePath)
    df=pd.DataFrame(data,columns=['x','y'])
    return df
```

```python
class1=read_dataset('class1.txt')
class2=read_dataset('class2.txt')
class3=read_dataset('class3.txt')
```

```python
print(class1.head())
print(class2.head())
print(class3.head())
```

```python
fig=plt.figure()
ax=fig.add_subplot()
ax.scatter(x=class1['x'],y=class1['y'],c='blue',marker='o')
ax.scatter(x=class2['x'],y=class2['y'],c='red',marker='x')
ax.scatter(x=class3['x'],y=class3['y'],c='green',marker='v')
```

```python
def split(df,savingpathtrain,savingpathtest):
    train=df.sample(frac=0.7)
    test=df.drop(train.index,inplace=False)
    train.to_csv(savingpathtrain)
    test.to_csv(savingpathtest)
    return train,test
```

```python
train_class1,test_class1=split(class1,'class1Train.csv','class1Test.csv')
train_class2,test_class2=split(class2,'class2Train.csv','class2Test.csv')
train_class3,test_class3=split(class3,'class3Train.csv','class3Test.csv')
```
